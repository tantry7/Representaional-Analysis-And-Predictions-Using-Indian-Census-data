## Year vs Total

```python
import seaborn as sns
%matplotlib inline
sns.pairplot(df,x_vars='Year',y_vars='Total',size=10,aspect=0.9,kind='reg')
```

```python
Year_var = df[['Year']]
Total_HDI = df[['Total']]
```

```python
print(Year_var)
```

```python
#Fitting the Linear Regression model
from sklearn.linear_model import LinearRegression
slr = LinearRegression()  
slr.fit(Year_var,Total_HDI)
```

```python
#new_years = pd.DataFrame(np.array([2020,2021,2022,2023,2025]).reshape(-1,1))
new_years = np.array([2020,2021,2022,2023,2024,2025]).reshape(-1,1)
```

```python
#Intercept and Coefficient
print("Intercept: ", slr.intercept_)
print("Coefficient: ", slr.coef_)
```

```python
y_pred_slr= slr.predict(new_years)
```

```python
y_pred_slr
```

```python
a = slr.intercept_ + (slr.coef_[0])*2020
```

```python
a
```

```python
from sklearn.metrics import mean_squared_error
import math
slrmse = mean_squared_error(Total_HDI.iloc[:len(new_years),:], y_pred_slr)
slrrmse = math.sqrt(mean_squared_error(Total_HDI.iloc[:len(new_years),:], y_pred_slr))
print(slrmse)
print(slrrmse)
```

```python
 #Fitting the Polynomial regression to the dataset  
from sklearn.preprocessing import PolynomialFeatures  
poly_regs= PolynomialFeatures(degree= 2)  
x_poly= poly_regs.fit_transform(Year_var)  
lin_reg_2 =LinearRegression()  
lin_reg_2.fit(x_poly, Total_HDI)  
```

```python
new_years_x_poly= poly_regs.fit_transform(new_years) 
```

```python
y_poly_pred = lin_reg_2.predict(new_years_x_poly)
```

```python
y_poly_pred
```

```python
from sklearn.metrics import mean_squared_error
import math
dmse = mean_squared_error(Total_HDI.iloc[:len(new_years),:], y_poly_pred)
drmse  = math.sqrt(mean_squared_error(Total_HDI.iloc[:len(new_years),:], y_poly_pred))
print(dmse)
print(drmse)
```

```python
#np.column_stack((df['Total'],y_poly_pred))
```

```python
# Fitting Decision Tree Regression to the dataset
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor.fit(Year_var,Total_HDI)
```

```python
y_pred_DTregressor= regressor.predict(new_years)
```

```python
y_pred_DTregressor
```

```python
from sklearn.metrics import mean_squared_error
import math
dmse = mean_squared_error(Total_HDI.iloc[:len(new_years),:], y_pred_DTregressor)
drmse  = math.sqrt(mean_squared_error(Total_HDI.iloc[:len(new_years),:], y_pred_DTregressor))
print(dmse)
print(drmse)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_regressor = RandomForestRegressor(n_estimators=10,random_state=0)
rf_regressor.fit(Year_var,Total_HDI)
y_pred_rf_regressor= rf_regressor.predict(new_years)
```

```python
y_pred_rf_regressor
```

```python
from sklearn.svm import SVR
SVM_regressor = SVR(kernel = 'rbf')
SVM_regressor.fit(Year_var, Total_HDI)
```

```python
y_pred_SVM = SVM_regressor.predict(new_years)
y_pred_SVM
```

```python
from sklearn.metrics import mean_squared_error
import math
smse =mean_squared_error(Total_HDI.iloc[:len(new_years),:], y_pred_SVM)
srmse=math.sqrt(mean_squared_error(Total_HDI.iloc[:len(new_years),:], y_pred_SVM))
print(smse)
print(srmse)
```

```python
from sklearn.linear_model import Ridge
rr = Ridge(alpha=0.01)
rr.fit(Year_var, Total_HDI)
y_pred_ridge = rr.predict(new_years)
```

```python
y_pred_ridge
```

```python
from sklearn.metrics import mean_squared_error
import math
rmse = mean_squared_error(Total_HDI.iloc[:len(new_years),:], y_pred_ridge)
rrmse =math.sqrt(mean_squared_error(Total_HDI.iloc[:len(new_years),:], y_pred_ridge))
print(rmse)
print(math.sqrt(mean_squared_error(Total_HDI.iloc[:len(new_years),:], y_pred_ridge)))
```

```python
from sklearn.linear_model import Lasso
lasso = Lasso()
lasso.fit(Year_var,Total_HDI)
y_pred_Lasso = lasso.predict(new_years)
```

```python
y_pred_Lasso
```

```python
from sklearn.metrics import mean_squared_error
import math
lmse = mean_squared_error(Total_HDI.iloc[:len(new_years),:], y_pred_Lasso)
lrmse = math.sqrt(mean_squared_error(Total_HDI.iloc[:len(new_years),:], y_pred_Lasso))
print(lmse)
print(lrmse)
```
