```python
years = [1991,2001,2003]
Total = [int(drinking_water_total_T['Total'][-1:]),int(drinking_water_total_T['Total.1'][-1:]),int(drinking_water_total_T['Total.2'][-1:])]
Total
```

```python
years = pd.DataFrame(years)
DW = pd.DataFrame(Total)
```

```python
years = years.rename(columns = {0:"Year"})
DW = DW.rename(columns ={0:"NGR"})
```

```python
#Fitting the Linear Regression model
from sklearn.linear_model import LinearRegression
slr = LinearRegression()  
slr.fit(years,DW)
```

```python
new_years = np.array([2021,2031]).reshape(-1,1)
```

```python
#Fitting the Linear Regression model
from sklearn.linear_model import LinearRegression
slr = LinearRegression()  
slr.fit(years,DW)
```

```python
print("Intercept: ", slr.intercept_)
print("Coefficient: ", slr.coef_)
```

```python
y_pred_slr= slr.predict(new_years)
```

```python
from sklearn.metrics import mean_squared_error
import math
slrmse = mean_squared_error(DW.iloc[:len(new_years),:], y_pred_slr)
slrrmse = math.sqrt(mean_squared_error(DW.iloc[:len(new_years),:], y_pred_slr))
print(slrmse)
print(slrrmse)
```

```python
 #Fitting the Polynomial regression to the dataset  
from sklearn.preprocessing import PolynomialFeatures  
poly_regs= PolynomialFeatures(degree= 2)  
x_poly= poly_regs.fit_transform(years)  
lin_reg_2 =LinearRegression()  
lin_reg_2.fit(x_poly, DW)  
```

```python
new_years_x_poly= poly_regs.fit_transform(new_years) 
```

```python
y_poly_pred = lin_reg_2.predict(new_years_x_poly)
```

```python
from sklearn.metrics import mean_squared_error
import math
pmse = mean_squared_error(DW.iloc[:len(new_years),:], y_poly_pred)
prmse =math.sqrt(mean_squared_error(DW.iloc[:len(new_years),:], y_poly_pred)) 
print(pmse)
print(prmse)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor.fit(years,DW)
```

```python
y_pred_DTregressor= regressor.predict(new_years)
```

```python
from sklearn.metrics import mean_squared_error
import math
dmse = mean_squared_error(DW.iloc[:len(new_years),:], y_pred_DTregressor)
drmse  = math.sqrt(mean_squared_error(DW.iloc[:len(new_years),:], y_pred_DTregressor))
print(dmse)
print(drmse)
```

```python
from sklearn.svm import SVR
SVM_regressor = SVR(kernel = 'rbf')
SVM_regressor.fit(years, DW)
```

```python
y_pred_SVM = SVM_regressor.predict(new_years)
```

```python
from sklearn.metrics import mean_squared_error
import math
smse =mean_squared_error(DW.iloc[:len(new_years),:], y_pred_SVM)
srmse=math.sqrt(mean_squared_error(DW.iloc[:len(new_years),:], y_pred_SVM))
print(smse)
print(srmse)
```

```python
from sklearn.linear_model import Ridge
rng = np.random.RandomState(0)
y = DW
X = years
rdg = Ridge(alpha = 0.2)
rdg.fit(X, y)
rdg.score(X,y)
ridgepredictor = rdg.predict(new_years)
```

```python
from sklearn.metrics import mean_squared_error
import math
rmse = mean_squared_error(DW.iloc[:len(new_years),:], ridgepredictor)
rrmse =math.sqrt(mean_squared_error(DW.iloc[:len(new_years),:], ridgepredictor))
print(rmse)
print(math.sqrt(mean_squared_error(DW.iloc[:len(new_years),:], ridgepredictor)))
```

```python

from sklearn.linear_model import Lasso
lasso = Lasso(alpha=1.0)
lasso.fit(years,DW)
LassoPredicter = lasso.predict(new_years)
```

```python
from sklearn.metrics import mean_squared_error
import math
lmse = mean_squared_error(DW.iloc[:len(new_years),:], LassoPredicter)
lrmse = math.sqrt(mean_squared_error(DW.iloc[:len(new_years),:], LassoPredicter))
print(lmse)
print(lrmse)
```
