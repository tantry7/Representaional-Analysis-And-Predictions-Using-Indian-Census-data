PREDICTION USING MACHINE LEARNING 


```python
year = list(Annual_birth_rate['Year'])
years =[]
for string in year:
    new_string = string.replace("*", "")
    years.append(new_string)
```

```python
for i in range(0, len(years)):
    years[i] = int(years[i])
```

```python
birth_rate = list(Annual_birth_rate.iloc[:,:3]['Total'])
for i in range(0, len(birth_rate)):
    birth_rate[i] = int(birth_rate[i])

```

```python
years = pd.DataFrame(years)
Birth_rate = pd.DataFrame(birth_rate)
```

```python
years = years.rename(columns = {0:"Year"})
```

```python
Birth_rate = Birth_rate.rename(columns ={0:"BR"})
```

```python
print(years)
print(Birth_rate)
```

```python
#Fitting the Linear Regression model
from sklearn.linear_model import LinearRegression
slr = LinearRegression()  
slr.fit(years,Birth_rate)
```

```python
#new_years = pd.DataFrame(np.array([2020,2021,2022,2023,2025]).reshape(-1,1))
new_years = np.array([2013,2014,2015,2016,2017,2018,2019,2020,2021,2022]).reshape(-1,1)
```

```python
#Intercept and Coefficient
print("Intercept: ", slr.intercept_)
print("Coefficient: ", slr.coef_)
```

```python
y_pred_slr= slr.predict(new_years)
```

```python
y_pred_slr
```

```python
a = slr.intercept_ + (slr.coef_[0])*2020
```

```python
a
```

```python
from sklearn.metrics import mean_squared_error
import math
slrmse = mean_squared_error(Birth_rate.iloc[:len(new_years),:], y_pred_slr)
slrrmse = math.sqrt(mean_squared_error(Birth_rate.iloc[:len(new_years),:], y_pred_slr))
print(slrmse)
print(slrrmse)
```

```python
 #Fitting the Polynomial regression to the dataset  
from sklearn.preprocessing import PolynomialFeatures  
poly_regs= PolynomialFeatures(degree= 2)  
x_poly= poly_regs.fit_transform(years)  
lin_reg_2 =LinearRegression()  
lin_reg_2.fit(x_poly, birth_rate)  
```

```python
new_years_x_poly= poly_regs.fit_transform(new_years) 
```

```python
y_poly_pred = lin_reg_2.predict(new_years_x_poly)
```

```python
y_poly_pred
```

```python
from sklearn.metrics import mean_squared_error
import math
dmse = mean_squared_error(Birth_rate.iloc[:len(new_years),:], y_poly_pred)
drmse  = math.sqrt(mean_squared_error(Birth_rate.iloc[:len(new_years),:], y_poly_pred))
print(dmse)
print(drmse)
```

```python
# Fitting Decision Tree Regression to the dataset
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor.fit(years,Birth_rate)
```

```python
y_pred_DTregressor= regressor.predict(new_years)
```

```python
y_pred_DTregressor
```

```python
from sklearn.metrics import mean_squared_error
import math
dmse = mean_squared_error(Birth_rate.iloc[:len(new_years),:], y_pred_DTregressor)
drmse  = math.sqrt(mean_squared_error(Birth_rate.iloc[:len(new_years),:], y_pred_DTregressor))
print(dmse)
print(drmse)
```

```python
from sklearn.svm import SVR
SVM_regressor = SVR(kernel = 'rbf')
SVM_regressor.fit(years, Birth_rate)
```

```python
y_pred_SVM = SVM_regressor.predict(new_years)
y_pred_SVM
```

```python
from sklearn.metrics import mean_squared_error
import math
smse =mean_squared_error(Birth_rate.iloc[:len(new_years),:], y_pred_SVM)
srmse=math.sqrt(mean_squared_error(Birth_rate.iloc[:len(new_years),:], y_pred_SVM))
print(smse)
print(srmse)
```

```python
from sklearn.linear_model import Ridge
rng = np.random.RandomState(0)

rdg = Ridge(alpha = 0.2)
rdg.fit(years, Birth_rate)
#rdg.score(X,y)
ridgepredictor = rdg.predict(new_years)
```

```python
from sklearn.metrics import mean_squared_error
import math
rmse = mean_squared_error(Birth_rate.iloc[:len(new_years),:], ridgepredictor)
rrmse =math.sqrt(mean_squared_error(Birth_rate.iloc[:len(new_years),:], ridgepredictor))
print(rmse)
print(math.sqrt(mean_squared_error(Birth_rate.iloc[:len(new_years),:], ridgepredictor)))
```

```python
from sklearn.linear_model import Lasso
lasso = Lasso(alpha=1.0)
lasso.fit(years,Birth_rate)
LassoPredicter = lasso.predict(new_years)
```

```python
from sklearn.metrics import mean_squared_error
import math
lmse = mean_squared_error(Birth_rate.iloc[:len(new_years),:], LassoPredicter)
lrmse = math.sqrt(mean_squared_error(Birth_rate.iloc[:len(new_years),:], LassoPredicter))
print(lmse)
print(lrmse)
```
